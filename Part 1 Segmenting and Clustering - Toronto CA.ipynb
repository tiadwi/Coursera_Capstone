### 1. Reading the Table
Use the Notebook to build the code to scrape the following Wikipedia page, https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M, in order to obtain the data that is in the table of postal codes and to transform the data into a pandas dataframe

```python
import pandas as pd

wiki_url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
read_table = pd.read_html(wiki_url,header=[0])
df = read_table[0]
df.head(10)
```

```python
df.shape
```

### 2. Renaming Data Frame 

```python
df = df.rename(columns={"Postal Code":"Postalcode", "Neighbourhood":"Neighborhood"}, index=str)
df.head()
```

### 3. Ignore cells with a borough that is Not assigned. 
Only process the cells that have an assigned borough. 

```python
df = df[df.Borough != 'Not assigned']
df.reset_index(drop=True, inplace=True)
df.head()
```

### 4. Grouping neighborhood based on same Postcalcode


```python
df = df.groupby("Postalcode", as_index=False).agg(lambda x: ','.join(set(x.dropna())))
df.head()
```

```python
df.shape
```

### 5. If a cell has a borough but a Not assigned neighborhood, then the neighborhood will be the same as the borough

```python
df.loc[df['Neighborhood'] == 'Not assigned','Neighborhood'] = df['Borough']
df.head()
```
